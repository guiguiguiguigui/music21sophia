```python
from music21 import *
```

```python
fp = '../new_out/enOrganized/lyrics_and_chords/wikifonia-1003.mxl'
```

```python
score = converter.parse(fp)
score.show()
```

```python
lastCs = None
lastCsOffset = 0
for cs in score.recurse().getElementsByClass('ChordSymbol'):
    thisCsOffset = cs.getOffsetInHierarchy(score)
    if lastCs is not None:
        lastCs.duration.quarterLength = thisCsOffset - lastCsOffset
    print(lastCs, thisCsOffset - lastCsOffset)
    lastCsOffset = thisCsOffset
    lastCs = cs
```

```python
chords = score.chordify()
notes = score.flat.getElementsByClass('Note').stream()
```

```python
offsetLyricDict = {}
for n in notes:
    flatOffset = n.getOffsetInHierarchy(score)
    offsetLyricDict[flatOffset] = n.lyric
```

```python
offsetLyricDict
```

```python
for c in chords.recurse().getElementsByClass('Chord'):
    c.closedPosition(forceOctave=4, inPlace=True)
    flatOffset = c.getOffsetInHierarchy(chords)
    if flatOffset in offsetLyricDict:
        c.lyric = offsetLyricDict[flatOffset]
    print(c, flatOffset, c.lyric)
```

```python
score.show()
```

```python
chords.show('text')
```
